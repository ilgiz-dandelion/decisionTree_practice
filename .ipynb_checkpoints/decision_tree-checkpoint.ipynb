### Decision trees

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('iris.csv', sep = ',')
df.head()
```

```python
iris = df.copy()
```

```python
iris['Species'].unique()
```

```python
iris['Species'] = iris['Species'].map({'Iris-setosa': 0,
                                      'Iris-versicolor': 1,
                                      'Iris-virginica': 2})
```

```python
iris.head()
```

```python
plt.scatter(x = iris['PetalLengthCm'], y = iris['PetalWidthCm'], c = iris['Species']);
```

```python
X = iris[['PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 99)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
model = DecisionTreeClassifier(criterion = "entropy")
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
plot_tree(model);
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,20))
_ = plot_tree(model, feature_names=['PetalLengthCm', 'PetalWidthCm'],
              class_names= ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], filled=True)
```

### Telecom Churn

```python
df = pd.read_excel('telecom_churn.xlsx')
df.head()
```

```python
df.pop('State')
df['International plan'] = df['International plan'].map({'Yes': 1, 'No': 0})
df['Voice mail plan'] = df['Voice mail plan'].map({'Yes': 1, 'No': 0})
```

```python
X = df.copy()
y = X.pop('Churn')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter = 10000)
lr.fit(X_train, y_train)
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
model = DecisionTreeClassifier(max_depth = 7)
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
print("Score of Tree on train:", model.score(X_train, y_train))
print("Score of Tree on test:", model.score(X_test, y_test))
print("Score of LogReg on train:", lr.score(X_train, y_train))
print("Score of LogReg on test:", lr.score(X_test, y_test))
```

```python
X_columns = X.columns.tolist()
```

```python
fig = plt.figure(figsize=(50,40))
_ = plot_tree(model, feature_names=X_columns,
              class_names=['True','False'], filled=True)
```
